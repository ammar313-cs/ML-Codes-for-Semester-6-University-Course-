```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(-2,2,100)
y = x**2

fig = plt.figure(figsize = (12,7))

plt.plot(x,y, alpha = 0.4, label =  'Y = X^2',
         color = 'blue', linestyle = 'dashed',
         linewidth = 10, marker = 'o',
         markersize = 5, markerfacecolor = 'red',
         markeredgecolor = 'blue')

plt.title("Eq plot")

plt.xlabel('x axis')
plt.ylabel('y axis')


fig.text(0.9, 0.15, 'graph',
         fontsize = 12, color ='red',
         ha ='right', va ='bottom',
         alpha = 0.7)

plt.grid(alpha = 1, linestyle = '--')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig1 = plt.figure(figsize = (14,8))

smaples = np.random.randn(10000)
plt.hist(smaples,bins = 30, density = True,
        alpha = 0.5, color ='red')
plt.title('plot title')

plt.xlabel('x axis')
plt.ylabel('Freq')

x1 = np.linspace(-4,4,100)
y1 = 1/(2 * np.pi)**0.5 * np.exp(-x1**2 / 2)

plt.plot(x1, y1, 'b', alpha = 0.8,color = 'blue')
plt.show()


```
